```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_theme(style='white', palette='Set2')
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
```

#### Define helper functions

```python
def get_sdt(df):
    hits = len(df[(df['Condition'] == 'TRUE') & (df['Resp'] == 'g')])
    misses = len(df[(df['Condition'] == 'TRUE') & (df['Resp'] == 'h')])
    correct_rejections = len(df[(df['Condition'] != 'TRUE') & (df['Resp'] == 'h')])
    false_alarms = len(df[((df['Condition'] != 'TRUE') & (df['Resp'] == 'g'))])

    hit_rate = hits / (hits+misses)
    fa_rate = false_alarms / (false_alarms+correct_rejections)

    z_hitRate = stats.norm.ppf(hit_rate)
    z_falseAlarmRate = stats.norm.ppf(fa_rate)
    d_prime = z_hitRate - z_falseAlarmRate

    return pd.Series({'Hit Rate':hit_rate, 'False Alarm Rate':fa_rate, 'd prime':d_prime})


def calculate_accuracy(df):
    hits = len(df[(df['Condition'] == 'TRUE') & (df['Resp'] == 'g')])
    correct_rejections = len(df[(df['Condition'] != 'TRUE') & (df['Resp'] == 'h')])

    return (hits+correct_rejections)/len(df)
```

#### Make DataFrames with participant responses and accuracy

```python
lst_df = []
for dir in (glob.glob(f'./data/responses/*')):
    dir_len = len(os.listdir(dir))
    if dir_len != 0:
        files = glob.glob(dir+'/*.csv')
        for i in range(len(files)):
            if len(files) == 4:
                if i != 0:
                    temp_df = pd.read_csv(files[i])
                    lst_df.append(temp_df)
            else:
                temp_df = pd.read_csv(files[i])
                lst_df.append(temp_df)

df = pd.concat(lst_df)
df['CorrectResp'] = df['Condition'].apply(lambda x: 'g' if x=='TRUE' else 'h') 
df['Accuracy'] = (df['CorrectResp']==df['Resp']).astype(int)
df['DisplayedDyad'] = df['AudioPath'].str.split('/').str[4]
df[['Age','Sex']] = df[['Sex','Age']].where(df['Subject'] > 5, df[['Age','Sex']].values)
df = df.reset_index(drop=True)
df
```

```python
df_accuracy = df.groupby('Subject', as_index=False).apply(lambda x: pd.Series({'Accuracy': calculate_accuracy(x)}))
df_accuracy
```

```python
sns.boxplot(y='Accuracy', data=df_accuracy, fill=False, width=0.2)
sns.stripplot(y='Accuracy', data=df_accuracy, alpha=0.3)
plt.xticks([])

res_acc = stats.ttest_1samp(df_accuracy['Accuracy'].to_numpy(), 0.5)
print(f't-test for above chance accuracy: t({res_acc.df})={res_acc.statistic}, p={res_acc.pvalue}')
```

#### t-tests for difference between d' of modalities

```python
df_sdt = df.groupby(['Subject', 'Block'], as_index=False).apply(lambda x: get_sdt(x)).melt(id_vars=['Subject', 'Block'], var_name='Measure').reset_index(drop=True)

ax = sns.boxplot(y='value', x='Measure', hue='Block', data=df_sdt, fill=False, gap=.1)
ax = sns.stripplot(y='value', x='Measure', hue='Block', data=df_sdt, alpha=0.3, dodge=True, legend=False)
ax.legend_.set_title('Modality')
for t, l in zip(ax.legend_.texts, ['A -> V', 'V -> V', 'AV -> V']):
    t.set_text(l)
# sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel('Measure')
plt.ylabel('Value')

vv_av = stats.ttest_rel(  df_sdt[(df_sdt['Measure']=='d prime') & (df_sdt['Block']=='vv')]['value'], 
                        df_sdt[(df_sdt['Measure']=='d prime') & (df_sdt['Block']=='va')]['value'])
print(f't-test between V -> V and A -> V dprime: t({vv_av.df})={vv_av.statistic}, p={vv_av.pvalue}')

vv_avv = stats.ttest_rel(  df_sdt[(df_sdt['Measure']=='d prime') & (df_sdt['Block']=='vv')]['value'], 
                        df_sdt[(df_sdt['Measure']=='d prime') & (df_sdt['Block']=='vva')]['value'])
print(f't-test between V -> V and AV -> V dprime: t({vv_avv.df})={vv_avv.statistic}, p={vv_avv.pvalue}')

av_avv = stats.ttest_rel(  df_sdt[(df_sdt['Measure']=='d prime') & (df_sdt['Block']=='va')]['value'], 
                        df_sdt[(df_sdt['Measure']=='d prime') & (df_sdt['Block']=='vva')]['value'])
print(f't-test between A -> V and AV -> V dprime: t({av_avv.df})={av_avv.statistic}, p={av_avv.pvalue}')
```

```python
sns.boxplot(y='value', x='Measure', data=df_sdt[df_sdt['Measure']=='d prime'], fill=False, width=0.2)
plt.xticks([])
plt.xlabel('')
plt.ylabel("d'")

df_sdt[df_sdt['Measure']=='d prime'].groupby('Block')['value'].agg(['mean', 'std'])
```
